```python
import seaborn as sns   
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=True, ax=ax)

plt.xlabel('Columns')
plt.ylabel('Missing values')
plt.show()
```

```python
import numpy as np

books['missing_description'] = np.where(books['description'].isnull(), 1, 0)
books['book_age'] = 2024 - books['published_year']
```

```python
columns_of_interst = ['num_pages', 'average_rating', 'missing_description', 'book_age']

correlation_matrix = books[columns_of_interst].corr(method = 'spearman')

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'spearman correlation'}) 
heatmap.set_title('Correlation Heatmap')

plt.show()
```

```python
book_missing = books[~(books['description'].isna())&
      ~(books['description'].isna())&
      ~(books['average_rating']).isna()&
      ~(books['average_rating']).isna()
      ]
book_missing
```

```python
books_greater_than_25 = book_missing[book_missing['words_in_description']>=25]
```

```python
books_greater_than_25
```

```python
books_greater_than_25['title_and_subtitle'] = (
    np.where(books_greater_than_25['subtitle'].isna(), books_greater_than_25['title'],
            (books_greater_than_25[['title', 'subtitle']].astype(str).agg(":".join, axis=1))))
```

```python
books_greater_than_25[['title', 'subtitle', 'title_and_subtitle']]
```

```python
books_greater_than_25['tagged_description'] = books_greater_than_25[['isbn13', 'description']].astype(str).agg(": ".join, axis=1)
books_greater_than_25
```

```python
(
    books_greater_than_25.drop(['subtitle', 'missing_description', 'words_in_description', 'book_age'], axis=1)
    .to_csv('books_cleaned.csv', index=False)
)
```
